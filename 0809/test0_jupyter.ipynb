##グラフを描画

$y=\sin(\alpha x), \alpha = 1, 2, 3$

を描画. jupter のテスト

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
alphas = [1, 2, 3]
x = np.linspace(0, 2*np.pi, 100)
for alpha in alphas:
    plt.plot(x,np.sin(alpha*x),label=str(alpha))
plt.xlim(0,2*np.pi)
plt.ylim(-1,1)
plt.show()
```
